**Fonction alea**

```python

def alea():
    global x, y, z
    x = 171 * (x % 177) - 2 * (x // 177)
    y = 172 * (y % 176) - 35 * (y // 176)
    z = 170 * (z % 178) - 63 * (z // 178)
    if x < 0:
        x += 30269
    if y < 0:
        y += 30307
    if z < 0:
        z += 30323
    inter = ((x / 30269.0) + (y / 30307.0) + (z / 30323.0))
    return round(inter - int(inter),4)


print("\n\nLes germes utilisés pour la simulation:\n")
while True:
    x = int(input("Entrer le premier germe x: "))
    if x >= 1 and x <= 30000:
        break

while True:
    y = int(input("Entrer le premier germe y: "))
    if y >= 1 and y <= 30000:
        break

while True:
    z = int(input("Entrer le premier germe z: "))
    if z >= 1 and z <= 30000:
        break
print("\n\nLes résultats de notre simulation:\n")
print("\tgerme x\t\t||\tgerme y\t\t||\tgerme z\t\t||\talea\t\t")
print("-----------------------------------------------------------------------------------------------")
for _ in range(10):
    print(f"\t{x}\t\t||\t{y}\t\t||\t{z}\t\t||\t{alea()}\t\t")
```

**Le salaire Actuelle** de chacun des Employé actuel ainsi que **dernier salaire ** de chacun des retraités

```python

salaires = [] #contient les salaires actuels des employes
dernier_salaire = [] #contient les derniers salaires des retraités



def salaire(salaires,dernier_salaire):
    salaire_intervals = [[30000, 40000], [20000, 30000], [15000, 20000], [10000, 15000], [7500, 10000], [5000, 7500], [3000, 5000]]
    freq_salaire_cumul = [5, 15, 25, 40, 60, 80, 100]
    nbr= 0
    for _ in range(11000): #10000 employer actuele et 1000 retraités
        IX = alea()  # Génère un nombre aléatoire entre 0 et 1
        IY = alea()  # Génère un nombre aléatoire entre 0 et 1

        # Recherche de l'intervalle de salaire correspondant à la valeur x
        intervalle = None  # Initialisation de l'intervalle à None
        for j, freq in enumerate(freq_salaire_cumul):
            if IX < freq / 100 and freq < 40:
                intervalle = salaire_intervals[j]
                break
            if IX <= freq / 100 and freq > 40:
                intervalle = salaire_intervals[j]
                break

        if intervalle is not None:
            salaire_exact = intervalle[0] + IY * (intervalle[1] - intervalle[0])
            salaire_exact = round(salaire_exact)
        if nbr < 1000:
                dernier_salaire.append(salaire_exact)
                nbr += 1
        else:
                salaires.append(salaire_exact)
    return salaires,dernier_salaire

```

**Fonction des salaires Embauche des Anciens et retraitées**



```python
Embouche_salaire=[]
S_Emb_retraiter=[]
def salaire_emb_ANC(Embouche_salaire,S_Emb_retraiter,salaires,dernier_salaire ):
    salaire_Emb_interval = [[24000, 32000], [16000, 24000], [12000, 16000], [8000, 12000], [6000, 8000], [4000, 6000], [3000, 4000]]
    S_Emb_freq_cumul = [5, 10, 20, 40, 60, 80, 100]
    L = 0
    while L < 11000:  #10000 employer actuele et 1000 retraités
        IX = alea()  # Génère un nombre aléatoire entre 0 et 1

        # Recherche de l'intervalle de salaire correspondant à la valeur x
        for j, freq in enumerate(S_Emb_freq_cumul):
            if IX < freq / 100 and freq < 40:
                intervalle = salaire_Emb_interval[j]
                break
            if IX <= freq / 100 and freq > 40:
                intervalle = salaire_Emb_interval[j]
                break

        if intervalle is not None:
            IY = alea()  # Génère un autre nombre aléatoire pour l'interpolation linéaire

            salaire_embauche_exact = intervalle[0] + IY * (intervalle[1] - intervalle[0])

            salaire_embauche_exact = round(salaire_embauche_exact)
            if L<1000 and salaire_embauche_exact < dernier_salaire[L]:
                S_Emb_retraiter.append(salaire_embauche_exact)

            elif  salaire_embauche_exact < salaires[L-1000]:
                 Embouche_salaire.append(salaire_embauche_exact)
            L += 1
    return Embouche_salaire, S_Emb_retraiter
```

**Fonction des ages actuel**

```python

ages = [] #contient les ages actuels
def age(ages):
    age_intervals = [[53, 62], [41, 52], [31, 40], [21, 30]]
    freq_cumul = [30, 60, 80, 100]
    for _ in range(10000):
        IX = alea()  # Génère un nombre aléatoire entre 0 et 1
        IY = alea()  # Génère un autre nombre aléatoire pour l'interpolation linéaire

        # Recherche de l'intervalle d'âge correspondant à la valeur x
        for j, freq in enumerate(freq_cumul):
            if IX <= freq / 100:
                intervalle = age_intervals[j]
                break

        if intervalle is not None:

            age_exact = intervalle[0] + IY * (intervalle[1] - intervalle[0])

            age_exact = round(age_exact)

            ages.append(age_exact)
    return ages
```

```python

def recruter():
    IX = alea()  # Génère un nombre aléatoire entre 0 et 1

    # Calcul du nombre de niveaux recrutés en utilisant la loi uniforme avec intervalles ouverts et fermés
    recrute = 250 + 150 * IX

    recrute = round(recrute)

    return recrute
```


**Nouveaux retraités **

```python

def nouveaux_retraite(employes, age_retraite):
    nouveaux_retraites = []
    indexes = []

    for e in employes:
        if e['age_actuel'] >= age_retraite:
            DSAR = e['salaire_actuel']
            NAT = age_retraite - e['age_embauche']
            PR = round(((NAT * 2) / 100) * DSAR)

            retraite = {
                'Pension_Retraite': PR
            }
            nouveaux_retraites.append(retraite)
            indexes.append(employes.index(e))

    for index in sorted(indexes, reverse=True):
        employes.pop(index)

    return nouveaux_retraites


```

```python
import matplotlib.pyplot as plt

# Récupérer les moyennes des années scenario1
moyennes_annes = []
for j in range(11):
    Moyenne = [0, 0, 0, 0, 0, 0, 0]
    for i in range(40):
        Moyenne[0] += L[i]['TotEmp'][j] / 40
        Moyenne[1] += L[i]['TotRet'][j] / 40
        Moyenne[2] += L[i]['TotCotis'][j] / 40
        Moyenne[3] += L[i]['TotPens'][j] / 40
        Moyenne[4] += L[i]['Reserve'][j] / 40
        Moyenne[5] += L[i]['NouvRet'][j] / 40
        Moyenne[6] += L[i]['NouvRec'][j] / 40
    for i in range(len(Moyenne)):
        Moyenne[i] = round(Moyenne[i])
    moyennes_annes.append(Moyenne)

# Extraire les années qui nous intéressent
annees = range(2023, 2033)
moyennes_selectionnees = [moyennes_annes[i] for i in range(len(moyennes_annes)) if i + 2023 in annees]

# Transposer les moyennes pour obtenir les données par catégorie
categories = ['TotEmp', 'TotRet', 'TotCotis', 'TotPens', 'Reserve', 'NouvRet', 'NouvRec']
moyennes_transposees = list(map(list, zip(*moyennes_selectionnees)))

# Créer les graphiques pour chaque catégorie
plt.figure(figsize=(10, 6))
for i in range(len(categories)):
    plt.figure(figsize=(10, 6))
    plt.plot(annees, moyennes_transposees[i], marker='o')
    plt.xlabel('Années')
    plt.ylabel('Moyenne')
    plt.title('Évolution de {}'.format(categories[i]))
    plt.grid(True)
    plt.show()
```
